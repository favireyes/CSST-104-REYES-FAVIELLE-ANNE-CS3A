Task 1: Data Loading

- Load the hardwareStore.csv dataset into a pandas DataFrame.

```python
import pandas as pd
df = pd.read_csv('hardwareStore.csv')
```

Task 2: Initial Data Analysis:

- Display basic statistics of the dataset, including mean, median, and standard deviation of numeric columns.

- Identify the unique categories and their counts.

```python
statistics = df.describe()
print(statistics)

uniq_category = df['CATEGORY_NAME'].unique()
```

Task 3: Category Analysis:

- Create a bar chart to visualize the distribution of products across categories.

- Determine which category has the highest number of products.

```python
import matplotlib.pyplot as plt
import seaborn as sne


product_dist = df.groupby('CATEGORY_NAME')['PRODUCT_ID'].sum()
highest_number_product = product_dist.idxmax()
product_dist.plot(kind='bar', color = 'green', width = 0.8)
plt.title('Distribution of Products Across Categories')
plt.xticks(rotation=90)
plt.show()

print("\n")
print('The category with the highest number of product: ' + highest_number_product)
```

Task 4: Cost and Price Analysis:

- Calculate the average standard cost and average list price.
- Identify the product with the highest standard cost and the one with the highest list price.

```python
avg_standard_cost = df['STANDARD_COST'].mean()
avg_list_prices = df['LIST_PRICE'].mean()

average = df[['PRODUCT_NAME', 'STANDARD_COST', 'LIST_PRICE']]
average2 = average.loc[average['STANDARD_COST'].idxmax()][0]
average3 = average.loc[average['LIST_PRICE'].idxmax()][0]
```

Task 5: Location Analysis:

- Determine the location (city or state) with the highest quantity of products in stock.


```python
val = df[['CITY', 'QUANTITY', 'PRODUCT_NAME']]
product_name = val.loc[val['QUANTITY'].idxmax()]

print(product_name[2])
print(product_name[0])

city = df.groupby('CITY')['QUANTITY'].sum()
city_highest = city.plot(kind='line', color = 'green')
plt.title('Quantity of Products in Stock per City')
plt.xticks(rotation=90)
```
