1. Creating Dataset

```python
import pandas as pd

data = {
    'student_id': range(1,21),
    'gender': ['Male','Female','Male','Male','Female','Male','Female','Male','Female','Male','Male','Male','Female','Male','Female','Male','Male','Female','Male','Female'],
    'test_date': pd.date_range(start='2024-02-21', periods=20),
    'study_hours': [2,3,5,7,3,4,5,2,1,4,2,3,1,5,2,3,4,1,2,3],
    'test_score': [80,75,86,81,83,84,85,92,91,94,92,93,81,75,82,38,84,91,82,93],
    'attendance': ['Present', 'Absent','Present', 'Present','Present',
                   'Present','Present', 'Present','Present', 'Present',
                   'Present', 'Present','Present', 'Absent','Present',
                   'Present','Present', 'Present','Present', 'Present'],
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

2. Data Loading and Initial Examination

```python
df = pd.read_csv('student_performance.csv')

print(df.head())

print("\n")

print(df.isnull().sum())
```

3. Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

# compute summary statistics
summary_stats = df[['test_score', 'study_hours']].describe()

print(summary_stats)

print("\n")

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'], bins=15, kde=True)
plt.title('Distribution of Test Score')
plt.xlabel('Test Scores')

plt.subplot(1,2,2)
sne.histplot(df['study_hours'], bins=15, kde=True)
plt.title('Distribution of Study Hours')
plt.xlabel('Study Hours')

plt.figure(figsize=(8,6))

sne.scatterplot(data=df, x='study_hours', y='test_score')
plt.title('Scatter Plot of Study Hours vs Test Score')
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
```

4. Data Visualization

```python
plt.figure(figsize=(6,4))
sne.countplot(data=df, x='gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')

plt.figure(figsize=(10,6))
sne.lineplot(data=df, x='test_date', y='test_score')
plt.title('Trend in Test Score Over Time')
plt.xlabel('Test Date')
plt.ylabel('Test Scores')
plt.xticks(rotation=45)

correlation_matrix = df[['study_hours','test_score','attendance']].corr()

plt.figure(figsize=(8,6))
sne.heatmap(correlation_matrix, annot=True, cmap='summer')
plt.title('Correlation Heatmap')
```
