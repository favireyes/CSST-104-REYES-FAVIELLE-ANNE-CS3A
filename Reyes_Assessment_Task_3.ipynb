# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2. Loading Dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')

print(data.head)
```

# 3. Preparing Data for Modeling

```python
X_exam= data[['Previous_Exam_Scores']] #IV
Y_exam = data[['Grades']] #DV
```

# 4. Fitting the Linear Regression Model

```python
model3 = LinearRegression()
model3.fit(X_exam,Y_exam)
```

# 5. Getting Model Coefficients

```python
slope3 = model3.coef_[0]
intercept3 = model3.intercept_

print('Slope: ', slope3)
print('Intercept: ', intercept3)
```

# 6. Visualizing the Linear Regression Line

```python
plt.scatter(X_exam,Y_exam, color = 'orange')
plt.plot(X_exam, model3.predict(X_exam), color = 'green')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variale (Y)')
plt.title('Linear Regression of Previous Exam Scores')
plt.show()
```

# 7. Making Predictions

```python
new_X = [[np.random.randint(100,200)]]
predicted_Y = model3.predict(new_X)

print('Value of X: ', new_X)
print('Predicted Value of Y: ', predicted_Y)
```
