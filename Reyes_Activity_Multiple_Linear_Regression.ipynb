# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# 2. Loading Dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset 2.csv')
print(data.head)
```

# 3. Preparing Data for Modeling

```python
X = data.drop('Grades', axis = 1)
Y = data['Grades']
```

#  4. Train and Split Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

# 5. Fitting the Linear Regression Model


```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

# 6. Making Predictions

```python
new_data = np.array([[5, 95, 90, 1, 30]])
predict_Y = model.predict(new_data)

grade = (predict_Y/60)*100

print(f"Predicted Grade: {grade[0]}")
```
